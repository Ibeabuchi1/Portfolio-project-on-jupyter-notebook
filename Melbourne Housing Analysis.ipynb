```python
#Examine the data

df = pd.read_csv(r'C:\Users\ACER-PC\Desktop\Z\csvs\melb_data.csv')
df.head(10)
```

```python
#describe the statistical nature of the data
df.describe()
```

```python
#check for Null values
df.isnull().sum(axis=0)
```

```python
#using the heat,ap to check the degree of null values present in the data
sns.heatmap(df.isnull(), cmap='viridis')
```

```python
#delete unusable columns

del df['BuildingArea']
del df['YearBuilt']
```

```python
#remove rows with null values

df2 = df
df2 = df2.dropna(subset=['CouncilArea'])
df2 = df2.dropna(subset=['Car'])
```

```python
#examine the data again

sns.heatmap(df2.isnull(), cmap='viridis')
```

```python
df2.head(20)
```

```python
# examine the data types of the columns

df2.dtypes
```

```python
# convert the floats into integers

df2['Price'] = df2['Price'].astype('int64')

df2['Distance'] = df2['Distance'].astype('int64')

df2['Postcode'] = df2['Postcode'].astype('int64')

df2['Bedroom2'] = df2['Bedroom2'].astype('int64')

df2['Bathroom'] = df2['Bathroom'].astype('int64')

df2['Car'] = df2['Car'].astype('int64')

df2['Landsize'] = df2['Landsize'].astype('int64')

df2['Propertycount'] = df2['Propertycount'].astype('int64')
```

```python
#Distinct count of the total regions

df2['Regionname'].drop_duplicates().sort_index(ascending=True)
```

```python
df2['Method'].drop_duplicates()
```

```python
#change the object type and get the last four digits of the date

df2['Year'] = df2['Date'].astype(str).str[-4:]
```

```python
# sort values according to price

df2.sort_values(by=['Price'], inplace=False, ascending=False)
```

```python
# to display all the results on the screen

pd.set_option('display.max_rows', 30)
```

```python
#check the correlation between variables
df.corr()
```

```python
#visualise the correlation
#methods are Kendall, Pearson and Spearman

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric datatype')

plt.xlabel('Housing features')

plt.ylabel('Housing features')

plt.show()
```

```python
#converting objects data type into category type for easy computation


df_numerics = df

for col_name in df_numerics.columns:
    if(df_numerics[col_name].dtypes == 'object'):
        df_numerics[col_name] = df_numerics[col_name].astype('category')
        df_numerics[col_name] = df_numerics[col_name].cat.codes
        
        
df_numerics.head(20)
    
```

```python
df_numerics.corr()
```

```python
#View the newly converted objects using seaborn matrix correlation matrix

correlation_matrix = df_numerics.corr(method='kendall')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric datatype')

plt.xlabel('Housing features')

plt.ylabel('Housing features')

plt.show()
```

```python
#pairing the correlated pairs

correlation_mat = df_numerics.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(ascending=True)

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)> 0.5].sort_values()

high_corr
```
